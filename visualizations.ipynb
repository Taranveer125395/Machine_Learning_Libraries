Matplotlib Library is used for creating data visualizations like different types of graphs.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
y2 = np.cos(x)
```

```python
plt.plot(x, y, color = 'g', linestyle = '-', label = 'Sine Wave')
plt.plot(x, y2, color = 'r', linestyle = '--', label = 'Cos Wave')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot Example')
plt.legend()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lineplot(x = 'size', y = 'total_bill', data = tips, hue = 'sex', style = 'sex', markers = True)
plt.title('Seaborn Line Chart')
plt.show()
```
