Matplotlib Library is used for creating data visualizations like different types of graphs.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
```

Line Plot

```python
plt.plot(x, y, color = 'g', linestyle = '-', label = 'Sine Wave')
plt.plot(x, y2, color = 'r', linestyle = '--', label = 'Cos Wave')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot Example')
plt.legend()
```

```python
plt.plot(x, y3, color = 'b', linestyle = ':', label = 'Tan Wave')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot Example')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lineplot(x = 'size', y = 'total_bill', data = tips, hue = 'sex', style = 'sex', markers = True)
plt.title('Seaborn Line Chart')
plt.show()
```

Bar Plot and Column Plot

```python
#plt.bar(tips['sex'], tips['total_bill'], color = 'orange', width = 0.5, label = 'Man Vs. Women')
plt.bar(tips['tip'], tips['total_bill'], color = 'orange', width = 0.5, label = 'Man Vs. Women')
plt.xlabel('Gender')
plt.ylabel('Total Bill')
plt.title('Total Bill Paid')
plt.legend()
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette = 'coolwarm')
plt.title('Seaborn Bar Chart')
```

Histogram

```python
data = np.random.randn(10000)
plt.hist(data, bins = 10, color = 'purple', edgecolor = 'black', alpha = 0.3)
```

Histogram with KDE

```python
sns.histplot(x = 'total_bill', data = tips, hue = 'sex', bins = 30, kde = True, color = 'blue')
```

```python
sns.histplot(tips['total_bill'], bins = 30, kde = True, color = 'blue')
```

Scatter Plot

```python
plt.scatter(tips['total_bill'], tips['tip'], label = 'Scatter plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Matplotlib Scatter Plot')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color = 'red', s = 50, alpha = 0.4, label = 'Data Points')
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', size = 'size', alpha = 0.8)
```

KDE Plot

```python
sns.kdeplot(x = 'total_bill',
            hue = 'sex',
            data = tips,
            fill = True,
            linewidth = 2,
            bw_adjust = 0.9)
```

Box Plot

```python
sns.boxplot(x = 'sex',
            y = 'total_bill',
            data = tips,
            palette = 'coolwarm')
```

```python
sns.boxplot(tips['total_bill'],
            palette = 'coolwarm')
```

Violin Plot

```python
sns.violinplot(x = 'sex',
               y = 'total_bill',
               data = tips,
               palette = 'Set2',
               inner = 'quartile')
```

Pie Chart

```python
categories = ['Category1',
              'Category2',
              'Category3',
              'Category4']
size = [30, 20, 15, 35]
color = ['#ff9999',
         '#66b3ff',
         '#99ff99',
         '#ffcc99']
plt.pie(size,
        labels = categories,
        autopct = '%1.1f%%',
        colors = color,
        startangle = 90)
```

Donut Chart

```python
plt.pie(size,
        labels = categories,
        autopct = '%1.1f%%',
        colors = color,
        startangle = 90)
centre_circle = plt.Circle((0, 0),
                           0.70,
                           fc = 'white')
plt.gca().add_artist(centre_circle)

plt.title('Donut Chart Example')
plt.show()
```
