Matplotlib Library is used for creating data visualizations like different types of graphs.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
```

```python
plt.plot(x, y, color = 'g', linestyle = '-', label = 'Sine Wave')
plt.plot(x, y2, color = 'r', linestyle = '--', label = 'Cos Wave')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot Example')
plt.legend()
```

```python
plt.plot(x, y3, color = 'b', linestyle = ':', label = 'Tan Wave')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot Example')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lineplot(x = 'size', y = 'total_bill', data = tips, hue = 'sex', style = 'sex', markers = True)
plt.title('Seaborn Line Chart')
plt.show()
```

```python
#plt.bar(tips['sex'], tips['total_bill'], color = 'orange', width = 0.5, label = 'Man Vs. Women')
plt.bar(tips['tip'], tips['total_bill'], color = 'orange', width = 0.5, label = 'Man Vs. Women')
plt.xlabel('Gender')
plt.ylabel('Total Bill')
plt.title('Total Bill Paid')
plt.legend()
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette = 'coolwarm')
plt.title('Seaborn Bar Chart')
```

```python
data = np.random.randn(10000)
plt.hist(data, bins = 10, color = 'purple', edgecolor = 'black', alpha = 0.3)
```

```python
sns.histplot(x = 'total_bill', data = tips, hue = 'sex', bins = 30, kde = True, color = 'blue')
```

```python
sns.histplot(tips['total_bill'], bins = 30, kde = True, color = 'blue')
```

```python
plt.scatter(tips['total_bill'], tips['tip'], label = 'Scatter plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Matplotlib Scatter Plot')
```
